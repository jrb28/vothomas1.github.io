```python
fig,ax = plt.subplots()
ax.scatter(df_opids['MedianHouseholdIncome'], df_opids['FPOO-Rate'], alpha=0.5)
fig.suptitle('Median Household Income vs Opioid overdoses')
ax.xaxis.set_label_text('Median Household Income in dollars')
ax.yaxis.set_label_text('Opioid overdoses')
fig.set_size_inches(7,5)
ax.set_xlim(ax.get_xlim()[::-1]) #Sets Median Household Income to descending order
plt.show()

#Scatterplots test correlation between two variables and in this case is Median Household Income and Opioid overdoses.

# Q: Do opioid overdoses tend to be associated with less affluent areasâ€”that is, areas where families have lower incomes? 

# A: Opioid overdoses do tend to be associated with areas of lower income as evidenced with the correlation below. 
```

```python
fig,ax = plt.subplots()
ax.scatter(df_opids['FPOO-Rate'], df_opids['FFHO-Rate'], alpha=0.5)
fig.suptitle('Opioid overdoses vs Fentanyl/Heroin overdoses')
ax.xaxis.set_label_text('Opioid overdoses')
ax.yaxis.set_label_text('Fentanyl/Heroin overdoses')
fig.set_size_inches(7,5)
plt.show()

#Scatterplots test correlation between two variables and in this case is Opioid overdoses and Fentanyl/Heroin overdoses.

# Q: Some people who start with opioid addictions are reported to transition to heroin use. 
#  : What is the relationship in Virginia counties between opioid overdoses and heroin overdoses?

# A: There appears to be some positive correlation between opioid overdoses and heroin overdoses. 
#  : However, whether opioid addiction translates to heroin use is unclear with the given data. 
```

```python
fig, ax = plt.subplots()
df_pres['State'].value_counts().plot(ax=ax, kind='bar')
fig.suptitle('Frequency of presidential birthplaces')
ax.xaxis.set_label_text('Birthplace')
ax.yaxis.set_label_text('Frequency')
fig.set_size_inches(7,5)
plt.show()

#Histograms allow variables to be compared based on frequency such as presidential birthplaces.

# Q: Which states are associated with the greatest number of United States presidents in terms of the presidentsâ€™ birthplaces?

# A: Virginia, Ohio, New York, and Massachusetts have the greatest number of United States presidents in terms of birthplace. 
#  : They have a frequency of 8, 7, 5, and 4, respectively. Vermont, Texas, and North Carolina have 3 while the remaining have 2.

```

```python
fig,ax = plt.subplots(figsize=(20, 10))
ax.plot(df_nsa['DATE'], df_nsa['TOTALNSA'], alpha=0.5)
ax.xaxis.set_major_locator(plt.MaxNLocator(10))
fig.suptitle('United States vehicle sales over time', fontsize=20)
ax.xaxis.set_label_text('Date', fontsize=20) #Professor, I would argue that this is not needed but is required based on the instructions
ax.yaxis.set_label_text('Total Monthly Sales', fontsize=20) 
plt.show()

# Line graphs are useful to display time-series data such as vehicle sales over time.

# Q: How have vehicle sales in the United States varied over time?

# A: Vehicle sales appear to be cyclical over a given year. Monthly sales appeared to generally increase until around 2008-2009.
#  : Since then, monthly sales appear to have recovered to levels experienced in 2000-2006.
```
